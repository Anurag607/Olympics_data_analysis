```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./athlete_events.csv")
df.shape
```

```python
# Id is compulsory field
df["ID"].count()
```

```python
team = "United States"
def country_participation(team):
    data = df.loc[df["Team"] == team].groupby(["Year"]).count()["ID"]
    plt.plot(data.index,data.values)
    plt.title(f"{team} Participation Over the Years")
    
country_participation("India")
```

```python
team = "United States"

def country_performance(team):
    for medal in ["Gold","Silver","Bronze"]:
        data = df.loc[(df["Team"] == team) & (df["Medal"] == medal)].groupby(["Year"]).count()["ID"]
        print(data)
        plt.plot(data.index,data.values, label=f"Total {medal} Medals")
    plt.title(f"{team} Performance Over the Years")
    plt.ylabel("No. of Medals")
    plt.legend()
    plt.xlabel("Year")
    
country_performance("India")
```

##### Medal Victory V/S Age

```python
for medal in ["Gold","Silver","Bronze"]:
    age_medal_data = df.loc[(df["Age"] > 22) & (df["Age"] < 38) & (df["Medal"] == medal)].groupby(["Age"]).count()["ID"]
    print(age_medal_data)
    plt.plot(age_medal_data.index, age_medal_data.values,label=medal)
plt.title("Medal Victory V/S Age")
plt.legend()
plt.show()
```

```python
for medal in ["Gold", "Silver", "Bronze"]:
    age_medal_data = df.query("Age > 22 and Age < 38 and Medal == @medal").groupby("Age").count()["ID"]
    plt.plot(age_medal_data.index, age_medal_data.values, label=medal)
plt.title("Medal Victory V/S Age")
plt.legend()
plt.show()
```
